```python
# Creating a new column.
df["Zeroes"] = ["Zero"for i in range(len(df))]
```

```python
# Use apply() for row-wise or column-wise custom calculations in DataFrames.
def product(a):
    return a *a

df["Double Salary"] = df["Salary"].apply(product)
```

```python
df["Double Salary"] = df["Double Salary"].astype(int)    # Changes Datatype of an column
```
