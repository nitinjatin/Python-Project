## Slicing.....

1-Dimensional Array Slicing with indices........

```python
array_1 = np.array([1,4,5,6,9,7,1,3,8,0])
print("Length : ",array_1.size)
print("Basic Slicing : ",array_1[1:6])
print("Basic Slicing with Steps : ",array_1[1:6:2])
print("Negative Slicing : ",array_1[-4:-1])
print("Negative Slicing with Steps : ",array_1[-1::-1])
```

2-Dimensional Array Slicing with indices........

```python
array_2 = np.array([[1,5,7,2],
                    [4,8,9,3],
                    [8,4,5,3],
                    [3,8,0,2]])
print("Specific Element : ",array_2[1,3])
print("Entire Row : ",array_2[1])
print("Entire Column : ",array_2[:,1])
```

## SORTING......

<!-- ðŸ§© Quick analogy:

axis=0	Act on rows, go down columns
axis=1	Act on columns, go across rows -->

```python
print("Soring in 1-d array : ",np.sort(array_1))
print("-----------------------------------------")
print("Soring in 2-d array \nOperates across the columns\nAffects the rows : ",np.sort(array_2,axis=1))                                  
print("-----------------------------------------")
print("Soring in 2-d array \nOperates down the rows\nAffects the columns : ",np.sort(array_2,axis=0))
print("-----------------------------------------")
```

## Filter.....

```python
# Boolean (conditional) indexing
# This is still indexing,
# but instead of saying â€œgive me element 1 and 2â€,
# youâ€™re saying â€œgive me all elements where this condition is True.â€

even_numbers = array_1[array_1%2 == 0]
print(even_numbers)
```

## Filter With Condition.....

```python
mask = array_1>5           # Creating a condition.
print(array_1)             
print(mask)                # Stores True or False after seeing the condition.
print(array_1[mask])       # Give me all elements where this condition is True.  
```

```python
# Creating a condition:
# where {1st arrugment} checks condition True or False.
# {2nd arrugment implements if condition is True}
# {3rd arrugment implements if condition is False}

condition_array = np.where(array_1>=5 , array_1*2 , array_1)
boolean_array = np.where(array_1>=5 , True , False)
print(condition_array)
print(boolean_array)
```

## Adding New Row And New Column

```python
original_array = np.array([[12,34,67],
                           [29,43,21],
                           [23,77,88]])
new_row = np.array([99,88,77])
with_new_row = np.vstack((original_array,new_row))
print("with_new_row: \n",with_new_row)
print("--------------------------------------------")

new_column = np.array([[10],[20],[30]])
with_new_column = np.hstack((original_array,new_column))
print("with_new_column: \n",with_new_column)
```
