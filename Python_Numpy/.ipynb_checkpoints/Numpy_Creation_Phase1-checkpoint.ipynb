```python
# flatten() Always returns a copy of the array.Changes made to the flattened array do NOT affect the original array.
# ravel() Returns a view of the array if possible (not always a copy).Changes made to the raveled array may reflect in the original array.

# ðŸ’¡ Rule of Thumb
# Use ravel() when you just need a temporary 1D view (faster, memory-friendly).
# Use flatten() when you explicitly need a separate copy (safe, independent).
```

## AXIS IN NUMPY........

```python
# axis=0: operates down the rows (affects columns)
# axis=1: operates across the columns (affects rows)

# ðŸ’¡ The key idea:
# â€œaxis=0 means operate down the rows,â€ that doesnâ€™t mean axis=0 is the columns â€”it means the operation moves along that axis.Same goes for 1.
```

```python
axis = np.array([[1,23,33,33],[12,33,45,55],[45,89,66,55]],np.int32)
```

```python
axis
```

```python
axis.shape
```

```python
axis.sum(axis=0)       # Moves or operates vertically â†“ (through rows)
```

```python
axis.sum(axis=1)      # Moves or operates horizontally â†’ (through columns)
```

```python
axis.argmin(axis=0)
```

```python
axis.argmin(axis=1)
```

```python
axis.argsort(axis=1)
```
