```python
import numpy as np
```

```python
# Data structure: [restaurant_id, 2020, 2022, 2023, 2024]
sales_data = np.array([
[1, 150000, 180000, 220000, 250000], # Paradise Biryani
[2, 120000, 140000, 160000, 190000], # Beijing Bites
[3, 200000, 230000, 260000, 300000], # Pizza Hub
[4, 180000, 210000, 240000, 270000], # Burger Point
[5, 160000, 185000, 205000, 230000] # Chai Point
])
```

## Accessing the Data

```python
print("-----Swiggy Sales Analysis------")
print("Shape of data" , sales_data.shape)
print("First Third Row of Sales Data: \n", sales_data[:4,:])
print("-----------------------------------------------------")
print("Only Sales from Sales Data: \n", sales_data[:,1:])
```

## Calculation

```python
print(sales_data)
print("Total Sales per Restaurent: ",np.sum(sales_data[:,1:],axis=1))
print("Total Sales per Year: ",np.sum(sales_data[:,1:],axis=0))
print("Maximum Sales per Restaurent: ",np.max(sales_data[:,1:],axis=1))
print("Minimum Sales per Year: ",np.min(sales_data[:,1:],axis=0))
print("Average Sales per Restaurent: ",np.mean(sales_data[:,1:],axis=1))
print("Monthly_Sales: \n", sales_data[:,1:]/12)      # Broadcasting , Scaler Element operates on every elements in an array.
```

## Vectorized Operation


```python
# NumPyâ€™s np.vectorize() takes a normal scalar function (like one that operates on a single value)
# and turns it into a vectorized function that can operate element-by-element on NumPy arrays. -->

vector = np.array(["Shirone","Amy","Pablo","Rian","Cecee"])
vectorized_array = np.vectorize(str.upper)
print("Vectorized Array: " , vectorized_array(vector))
```
